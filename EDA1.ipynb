## 1.	Data Cleaning and Preparation:

```python
df =pd.read_csv(r"C:\Users\mobin\DATASIENCE\ML Assignments\STATISTICS\EDA1\Cardiotocographic.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# finding the mean values for each column
# filling nun values with the mean
for i in df.columns:
    # Searching through every column
    mean_value = df[i].mean()
    print(mean_value)
    # if null avalible fill with the mean value
    if df[i].isnull().sum() != 0:
        df[i] = df[i].fillna(mean_value)
```

```python
df.isnull().sum()
```

```python
# OUTLIERS CHEACKING

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,8))
sns.boxenplot(df)
```

## 2. Statistical Summary

```python
# Provide a statistical summary for each variable in the dataset, including measures of central tendency (mean, median)
# dispersion (standard deviation, interquartile range).
import numpy as np
mean_list = []
median_list = []
std = []
for i in df.columns:
    mean = np.mean(df[i])
    meadian = np.median(df[i])
    standerd_dev = np.std(df[i])
    median_list.append(meadian)
    mean_list.append(mean)
    std.append(standerd_dev)

dataframe = pd.DataFrame(index=df.columns)
dataframe['Mean'] = mean_list
dataframe['Median'] = median_list
dataframe['Standered_deviation'] = std
```

## 3.Data Visualization:

```python
# Create histograms or boxplots to visualize the distributions of various numerical variables.

for i in df:
    plt.boxplot(df[i])
    plt.title(i)
    plt.show()
```

```python
for i in df:
    sns.histplot(df[i], kde = True,
                line_kws = {'linestyle':'dashed',
                         'linewidth':'2'}).lines[0].set_color('red')
    plt.title(i)
    plt.show()
```

```python
# Pie plot to show companies
A = []
B = []

for i in df.columns:
    mean_1 = np.mean(df[i])
    if mean_1 > 5:
        A.append(i)
        B.append(mean_1)
plt.pie(B,labels=A,autopct='%1.0f%%')
```

```python
# Generate scatter plots or correlation heatmaps to explore relationships between pairs of variables

sns.heatmap(data = df)
```
